## Import Semua Packages/Library yang Digunakan

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import streamlit as st
```

## Data Wrangling

### Gathering Data

```python
# Load the datasets
day_data = pd.read_csv('day.csv')
hour_data = pd.read_csv('hour.csv')
```

```python
# Tampilkan beberapa baris pertama dari masing-masing DataFrame
print("Day Data:")
print(day_data.head())
```

```python
print("Hour Data:")
print(hour_data.head())
```

**Insight:**
- Terdapat dua file dataset yang dikumpulkan yaitu day.csv dan hour .csv
- Diperoleh beberapa baris pertama dari masing masing dataframe

### Assessing Data

```python
# Tampilkan informasi ringkas tentang DataFrame
print(day_data.info())
print(hour_data.info())
```

```python
# Memeriksa duplikasi
day_duplicates = day_data.duplicated().sum()
hour_duplicates = hour_data.duplicated().sum()
print(f"Jumlah duplikasi di Day Data: {day_duplicates}")
print(f"Jumlah duplikasi di Hour Data: {hour_duplicates}")
```

```python
# Periksa nilai yang hilang
print("Missing values in Day Data:")
print(day_data.isnull().sum())
print("\nMissing values in Hour Data:")
print(hour_data.isnull().sum())
```

**Insight:**
- Diperoleh informasi ringkas dalam kedua dataset
- Tidak ada missing value maupun duplikasi data dalam kedua dataset

### Cleaning Data

```python
# Mengonversi kolom dteday ke tipe datetime
day_data['dteday'] = pd.to_datetime(day_data['dteday'])
hour_data['dteday'] = pd.to_datetime(hour_data['dteday'])
```

```python
# Menambahkan kolom hari dalam seminggu
day_data['weekday'] = day_data['dteday'].dt.day_name()
hour_data['weekday'] = hour_data['dteday'].dt.day_name()
```

**Insight:**
- Diperoleh data yang bersih dan siap untuk dianalisis
- Menambahkan kolom hari dalam seminggu untuk memperdalam analisis

## Exploratory Data Analysis (EDA)

### Explore ...

```python
### Univariate analysis
```

```python
# Histogram untuk jumlah peminjaman
plt.figure(figsize=(10, 6))
sns.histplot(hour_data['cnt'], bins=30, kde=True)
plt.title('Distribusi Jumlah Peminjaman Sepeda')
plt.xlabel('Jumlah Peminjaman')
plt.ylabel('Frekuensi')
plt.show()
```

```python
# Bar plot untuk frekuensi musim
plt.figure(figsize=(10, 6))
sns.countplot(data=day_data, x='season', palette='pastel')
plt.title('Frekuensi Peminjaman Berdasarkan Musim')
plt.xlabel('Musim')
plt.ylabel('Frekuensi')
plt.xticks(ticks=[0, 1, 2, 3], labels=['Musim Dingin', 'Musim Semi', 'Musim Panas', 'Musim Gugur'])
plt.show()
```

```python
### Bivariate Analysis
```

```python
# Scatter plot antara suhu dan jumlah peminjaman
plt.figure(figsize=(10, 6))
sns.scatterplot(data=hour_data, x='temp', y='cnt', alpha=0.6)
plt.title('Hubungan antara Suhu dan Jumlah Peminjaman')
plt.xlabel('Suhu (dalam skala 0-1)')
plt.ylabel('Jumlah Peminjaman')
plt.show()
```

```python
### Multivariate Analysis
```

```python
# Memilih hanya kolom numerik
numeric_data = hour_data.select_dtypes(include=['float64', 'int64'])

# Hitung matriks korelasi
correlation_matrix = numeric_data.corr()

# Heatmap untuk melihat korelasi antar variabel
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Korelasi Antar Variabel')
plt.show()
```

```python
### Time series analysis
```

```python
# Mengelompokkan data berdasarkan bulan
monthly_rentals = hour_data.groupby(hour_data['dteday'].dt.to_period('M'))['cnt'].sum()

plt.figure(figsize=(12, 6))
monthly_rentals.plot(kind='line')
plt.title('Tren Peminjaman Sepeda per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Peminjaman')
plt.grid()
plt.show()
```

**Insight:**
- Diperoleh beberapa analisis data untuk day data
- Diperoleh beberapa analisis data untuk hour data

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
# Visualisasi jumlah peminjaman sepeda berdasarkan jam
plt.figure(figsize=(12, 6))
sns.lineplot(data=hour_data, x='hr', y='cnt', marker='o', color='blue', linewidth=2, markersize=8)
plt.title('Jumlah Peminjaman Sepeda Berdasarkan Jam dalam Sehari', fontsize=16)
plt.xlabel('Jam', fontsize=14)
plt.ylabel('Jumlah Peminjaman', fontsize=14)
plt.xticks(range(0, 24))  # Menampilkan jam dari 0 hingga 23
plt.grid(color='gray', linestyle='--', linewidth=0.5)

# Menambahkan area di bawah garis untuk memberikan efek visual
plt.fill_between(hour_data['hr'], hour_data['cnt'], color='lightblue', alpha=0.3)

plt.show()
```

### Pertanyaan 2:

```python
# Menghitung jumlah peminjaman berdasarkan hari dalam seminggu
weekly_rentals = hour_data.groupby('weekday')['cnt'].sum().sort_index()

# Visualisasi tren peminjaman sepeda berdasarkan hari dalam seminggu (Diagram Batang Horizontal)
plt.figure(figsize=(12, 6))
ax = sns.barplot(y=weekly_rentals.index, x=weekly_rentals.values, palette='viridis')
plt.title('Jumlah Peminjaman Sepeda Berdasarkan Hari dalam Seminggu')
plt.xlabel('Jumlah Peminjaman')
plt.ylabel('Hari')
plt.grid(axis='x')

# Mengatur jarak antar tick pada sumbu X
plt.xticks(range(0, 500001, 25000))  # Menampilkan tick setiap 25.000

# Menambahkan label pada setiap batang dengan jarak
for p in ax.patches:
    ax.annotate(f'{int(p.get_width())}', (p.get_width(), p.get_y() + p.get_height() / 2), 
                ha='left', va='center', fontsize=10, color='black', 
                xytext=(5, 0), textcoords='offset points')  # Menambahkan offset

plt.xlim(0, 550000)  # Memberikan sedikit ruang di sebelah kanan
plt.xticks([])
plt.show()
```

**Insight:**
- Diperoleh hasil analisis dari pertanyaan bisnis 1
- Diperoleh hasil analisis dari pertanyaan bisnis 2

## Analisis Lanjutan (Opsional)

```python
# Menghitung jumlah peminjaman berdasarkan musim
seasonal_rentals = day_data.groupby('season')['cnt'].sum()

# Visualisasi
plt.figure(figsize=(8, 5))
sns.barplot(x=seasonal_rentals.index, y=seasonal_rentals.values, palette='pastel')
plt.title('Jumlah Peminjaman Sepeda Berdasarkan Musim')
plt.ylabel('Jumlah Peminjaman')
plt.xticks(ticks=[0, 1, 2, 3], labels=['Musim Dingin', 'Musim Semi', 'Musim Panas', 'Musim Gugur'])
plt.show()
```

## Conclusion

- Conclution pertanyaan 1
  Dari analisis yang dilakukan, terlihat bahwa jumlah peminjaman sepeda bervariasi sepanjang hari. Puncak peminjaman terjadi pada jam-jam tertentu, biasanya pada pagi hari (sekitar jam 7-9) dan sore hari (sekitar jam 17-19), yang menunjukkan bahwa banyak pengguna sepeda menggunakan layanan ini untuk berangkat dan pulang kerja atau sekolah. Sebaliknya, jumlah peminjaman cenderung lebih rendah pada malam hari dan dini hari, yang mungkin disebabkan oleh berkurangnya aktivitas di luar ruangan pada waktu-waktu tersebut.
  
- Conclution pertanyaan 2
  Analisis menunjukkan bahwa jumlah peminjaman sepeda bervariasi berdasarkan hari dalam seminggu. Biasanya, jumlah peminjaman tertinggi terjadi pada hari kerja (Senin hingga Jumat), sementara jumlah peminjaman cenderung lebih rendah pada akhir pekan (Sabtu dan Minggu). Hal ini mungkin mencerminkan pola penggunaan sepeda yang lebih tinggi di kalangan pekerja dan pelajar yang menggunakan sepeda sebagai sarana transportasi sehari-hari. Selain itu, faktor-faktor seperti cuaca, acara lokal, dan kebijakan transportasi juga dapat mempengaruhi pola peminjaman pada hari-hari tertentu.
